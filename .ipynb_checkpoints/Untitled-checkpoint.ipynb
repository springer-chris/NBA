```python
import pandas as pd
import numpy as np

kg = pd.read_csv("CSVs/Kaggle Data/NBA_PBP_2020-21.csv")
kg["play_count"] = 1
kg["season"]= "2020-2021"
kg['ShotPoints'] = kg['ShotType'].str[:4]
#Primary questions for basic understanding
## is there a consistent way to add an and/or in the middle of a function statement? 
# Yes. encircle each part with parenthesis and use normal logical functions like ==, !=
##cast or convert function
kg.assign(GameStartDateTime=lambda kg : pd.to_datetime(kg.Date + ' ' + kg.Time))
#Is there one way that works for all dtypes or not really?
#kg['columnname'].as_type('newdatatype')
### how to include multiple factors in a regex for example
##case when
condlist = (kg['FreeThrowOutcome']=='make',kg['ShotPoints']=='2-pt',kg['ShotPoints']=='3-pt')
vallist = (1,2,3)
kg['Points'] = np.select(condlist,vallist,default=0)
kg['ShotVersion'] = np.select(
[
    (kg['ShotPoints']=='2-pt'),(kg['ShotPoints']=='3-pt'),(pd.notnull(kg['FreeThrowOutcome']))
],
[    ('fg'),('3'),('ft')
],default='NA'
)

kg['ShotSuccess'] = np.select(
[
    ((kg['FreeThrowOutcome']=='make')|(kg['ShotOutcome']=='make')),((kg['FreeThrowOutcome']=='miss')|(kg['ShotOutcome']=='miss'))
],
[
    (1),(0)
],default='NaN'
)

kg['PlayType'] = np.select(
[
    
    (pd.notnull(kg['FreeThrowOutcome']))
     ,(pd.notnull(kg['ShotOutcome']))
     ,(pd.notnull(kg['Rebounder']))
     ,(pd.notnull(kg['FoulType']))
     ,(pd.notnull(kg['JumpballAwayPlayer']))
     ,(pd.notnull(kg['TurnoverType']))
     ,(pd.notnull(kg['EnterGame']))
     ,(pd.notnull(kg['TimeoutTeam']))
     ,(pd.notnull(kg['ViolationType']))
     ,((kg['AwayPlay'].str[:3]=='End')|(kg['HomePlay'].str[:3]=='End'))
     ,((kg['AwayPlay'].str[:7]=='Instant')|(kg['HomePlay'].str[:7]=='Instant'))
     ,((kg['AwayPlay'].str[:5]=='Def 3')|(kg['HomePlay'].str[:5]=='Def 3'))
    

],
[
    ('FT')
    ,('Shot')
    ,('Rebound')
    ,('Foul')
    ,('Jump Ball')
    ,('Turnover')
    ,('Sub')
    ,('Timeout')
    ,('Violation')
    ,('End of Q')
    ,('Replay')
    ,('Def 3 Sec')
    
],default='Other'
)

##where

kg['Points'] = kg['Points'].astype('float64')
## what is the closest equivalent to an in
## what is the closest equivalent to like '%%'
## what is the closest equivalent to a partition
## what is the closest equivalent to a group by, specficially by multiple elements
# what is the closest equivalent to a select distinct
##### for a single column use Unique()       ex df.column.unique()    use drop_duplicates() or all columns or specifcy the specic columns
## how to sort by multiple elements
##### df.sort_values(['a','b'],ascending=[True,False])
## how to check data types
si = kg.loc[kg['ShotVersion']!='NA',['FreeThrowOutcome','ShotOutcome','Points','ShotVersion','ShotSuccess']]

#sc = si[['Points','ShotVersion','ShotSuccess']].groupby(['ShotVersion','ShotSuccess']).count()
#sc = sc.rename(columns = {'Points':'CountOfShots'})

kg
#sc.loc[sc['ShotSuccess']=='make',['ShotVersion','Points']]
#t = kg.loc[pd.notnull(kg['TurnoverPlayer'])]
#jb = kg.loc[pd.notnull(kg['JumpballAwayPlayer'])]
#lg = kg.loc[pd.notnull(kg['LeaveGame'])]
#s = kg.loc[pd.notnull(kg['Shooter'])]
#ft = kg.loc[pd.notnull(kg['FreeThrowShooter'])]
#r = kg.loc[pd.notnull(kg['Rebounder'])]
#a = kg.loc[pd.notnull(kg['Assister'])]
#games = kg[['Date','Time','HomeTeam','AwayTeam']].drop_duplicates()

#kg.sort_values(['Date','Time'], ascending=[True,True])  #,'HomeTeam','AwayTeam','Quarter','SecLeft'])

#tfg = kg.groupby("TurnoverPlayer").rank(method='min') #,"Location","Date","Time","HomeTeam","AwayTeam")

```
